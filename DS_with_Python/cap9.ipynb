```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
data = np.random.rand(10)
print(data)

plt.plot(data)
```

```python
imagem = plt.figure()
ax1 = imagem.add_subplot(2,2,1)
ax2 = imagem.add_subplot(2,2,2)
ax3 = imagem.add_subplot(2,2,3)
```

```python
#exemplo de subplot

imagem = plt.figure()
ax1 = imagem.add_subplot(2,2,1)
plt.plot(np.random.randn(10).cumsum(), 'k-')
ax2 = imagem.add_subplot(2,2,2)
plt.plot(np.random.random(10).cumsum())
ax3 = imagem.add_subplot(2,2,3)
plt.plot(np.random.randn(10), 'k--')
```

```python
imagem = plt.figure()
ax1 = imagem.add_subplot(2,2,1)
plt.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
ax2 = imagem.add_subplot(2,2,2)
plt.hist(np.random.randn(50), bins=30, color='red')
ax3 = imagem.add_subplot(2,2,3)
plt.plot(np.random.randn(10), 'k--')
```

```python
plt.plot(np.random.randn(15),'ko--' )
```

```python
exem = np.random.randn(20).cumsum()

plt.plot(exem,'k--', label='default' )
plt.plot(exem, 'k-', drawstyle='steps-post', label='steps-post')
```

```python
import pandas as pd
new_way = pd.Series(np.arange(30).cumsum() + 3 * np.random.randn(30))
new_way.tail()
new_way.plot()
```

```python
new_way2 = pd.DataFrame(np.random.randn(10,4), 
    columns=['Aa', 'Bb', 'Cc', 'Dd'],
    index = np.arange(0,100,10) )

new_way2.cumsum().plot()
```

```python
imagem = plt.figure()
ax1 = imagem.add_subplot(2,2,1)
new_way2['Aa'].plot.bar()
ax2 = imagem.add_subplot(2,2,2)
new_way2['Aa'].cumsum().plot.bar()
ax3 = imagem.add_subplot(2,2,3)
new_way2.cumsum().plot.bar()
```

```python
new_way2.cumsum().plot.barh(stacked = True)
```
